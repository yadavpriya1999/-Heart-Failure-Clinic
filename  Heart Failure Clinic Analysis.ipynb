```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("heart_failure_clinical_records_dataset.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
#type casting
df['sex']=df['sex'].map(str)
df['anaemia']=df['anaemia'].map(str)
df['diabetes']=df['diabetes'].map(str)
df['high_blood_pressure']=df['high_blood_pressure'].map(str)
df['smoking']=df['smoking'].map(str)
```

```python
df.info()
```

```python
#describing categorical data and numerical data
```

```python
df['sex']=df['sex'].str.replace("0","female")
```

```python
df['sex']=df['sex'].str.replace("1","male")
```

```python
df.head()
```

# 1.What is the distribution of age among heart failure patients in the dataset.

```python
#Age Distribution
plt.title("Age Distribution")
plt.hist(df['age'],bins=20)
plt.show()
```

```python
sns.histplot(df['age'],kde=True, color="m")
```

# 2. How does the death rate vary with age.

```python
age_groups = df.groupby('age')['DEATH_EVENT'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(age_groups['age'], age_groups['DEATH_EVENT'], marker='o')
plt.xlabel('Age')
plt.ylabel('Death Rate')
plt.title('Death Rate Variation with Age')
plt.grid(True)
plt.show()
```

# 3.What is the percentage of male and female patients in the dataset?

```python
sns.countplot(data=df,x="sex")
```

```python
total_patients = len(df)
male_patients = df[df['sex'] == 'male']['sex'].count()
female_patients = df[df['sex'] == 'female']['sex'].count()

# Calculate the percentage of male and female patients
percentage_male = (male_patients / total_patients) * 100
percentage_female = (female_patients / total_patients) * 100

labels = ['Male', 'Female']
sizes = [percentage_male, percentage_female]
colors = ['lightblue', 'lightcoral']
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Percentage of Male and Female Patients')
plt.show()
```

# 4. How does platelet count vary among different age group.

```python
platelet_age_groups = df.groupby('age')['platelets'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(platelet_age_groups['age'], platelet_age_groups['platelets'], marker='o')
plt.xlabel('Age')
plt.ylabel('Average Platelet Count')
plt.title('Platelet Count Variation with Age')
plt.grid(True)
plt.show()
```

# 5. Is there any correlation between creatinine and sodium level in blood?

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='serum_creatinine', y='serum_sodium', data=df)
plt.xlabel('Serum Creatinine')
plt.ylabel('Serum Sodium')
plt.title('Creatinine vs Sodium Levels in Blood')
plt.show()

# Calculate Pearson correlation coefficient
correlation = df['serum_creatinine'].corr(df['serum_sodium'])
print(f"Pearson correlation coefficient: {correlation:.2f}")
```

# 6.How does the prevalence of high blood pressure differ between male and female patients?

```python
total_male = df[df['sex'] == 'male']['sex'].count()
total_female = df[df['sex'] == 'female']['sex'].count()

# Calculate the number of male and female patients with high blood pressure
male_hbp = df[(df['sex'] == 'male') & (df['high_blood_pressure'] == '1')]['sex'].count()
female_hbp = df[(df['sex'] == 'female') & (df['high_blood_pressure'] == '1')]['sex'].count()

# Calculate the percentage of male and female patients with high blood pressure
percentage_male_hbp = (male_hbp / total_male) * 100
percentage_female_hbp = (female_hbp / total_female) * 100

print(f"Percentage of male patients with high blood pressure: {percentage_male_hbp:.2f}%")
print(f"Percentage of female patients with high blood pressure: {percentage_female_hbp:.2f}%")

```

```python
plt.figure(figsize=(8, 6))
plt.bar(['Male', 'Female'], [male_hbp, female_hbp], color=['skyblue', 'lightcoral'])
plt.xlabel('Gender')
plt.ylabel('Percentage with High Blood Pressure')
plt.title('Prevalence of High Blood Pressure by Gender')
plt.ylim(0, 100)
plt.show()
```

# 7. what is the relationship between smoking habits and the occurrence of heart failure?

```python
smoking_hf = df[df['smoking'] == '1']['DEATH_EVENT'].mean() * 100
non_smoking_hf = df[df['smoking'] == '0']['DEATH_EVENT'].mean() * 100

plt.figure(figsize=(8, 6))
plt.bar(['Smoking', 'Non-Smoking'], [smoking_hf, non_smoking_hf], color=['skyblue', 'lightcoral'])
plt.xlabel('Smoking Habits')
plt.ylabel('Percentage with Heart Failure')
plt.title('Relationship Between Smoking Habits and Heart Failure')
plt.ylim(0, 100)
plt.show()
```

# 8. Are there any noticeable pattern in the distribution of death events across different age group?

```python
death_events_by_age = df.groupby(['age', 'DEATH_EVENT']).size().unstack(fill_value=0)
plt.figure(figsize=(12, 6))
plt.bar(death_events_by_age.index, death_events_by_age[1], color='red', label='Death Events')
plt.xlabel('Age')
plt.ylabel('Number of Death Events')
plt.title('Distribution of Death Events Across Different Age Groups')
plt.legend()
plt.show()
```

Insights: maximum death at age 60

# 9. Is there any significant difference in ejection fraction between patients with and without diabetes?

```python
ef_diabetes = df[df['diabetes'] == "1"]['ejection_fraction'].mean()
ef_no_diabetes = df[df['diabetes'] == "0"]['ejection_fraction'].mean()


plt.figure(figsize=(8, 6))
plt.bar(['Diabetes', 'No Diabetes'], [ef_diabetes, ef_no_diabetes], color=['skyblue', 'lightcoral'])
plt.xlabel('Diabetes Status')
plt.ylabel('Average Ejection Fraction')
plt.title('Difference in Ejection Fraction Between Patients with and Without Diabetes')
plt.ylim(0, 100)
plt.show()
```

# 10. How does the serum creatinine level vary between patients who survived and those who did not?

```python
creatinine_survived = df[df['DEATH_EVENT'] == 0]['serum_creatinine'].mean()
creatinine_not_survived = df[df['DEATH_EVENT'] == 1]['serum_creatinine'].mean()

plt.figure(figsize=(8, 6))
plt.bar(['Survived', 'Not Survived'], [creatinine_survived, creatinine_not_survived], color=['skyblue', 'lightcoral'])
plt.xlabel('Outcome')
plt.ylabel('Average Serum Creatinine')
plt.title('Variation of Serum Creatinine Between Survived and Not Survived Patients')
plt.ylim(0, 10)  
plt.show()
```
